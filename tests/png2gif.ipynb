```python
from PIL import Image
from tqdm.notebook import tqdm
import os
```

```python
input_directory = '../results/stack/figures/360'

output_file = '../results/360_dpi_100.gif'
```

```python
fps = 15

# Load all the JPG files from the directory
jpg_files = sorted([f for f in os.listdir(input_directory) if f.endswith('.jpg')])[::5]#[60*23:60*24:2]
```

```python
jpg_files
```

```python
len(jpg_files)
```

```python
# Create a list to hold the image frames
frames = []

resize_factor = 1

# Iterate over each JPG file

for ifile,file_name in tqdm(enumerate(jpg_files[:])):
    # Open the image file
    image = Image.open(os.path.join(input_directory, file_name))
    #image2 = Image.open(os.path.join('EPIC/figures/usgs/nc73797331_best_otime/', jpg_files_2[ifile]))
    # Convert the image to RGB mode only if it's not already in RGB
    if image.mode != 'RGBA':
        image = image.convert('RGBA')
    
    
    # Append the image to the frames list
    frames.append(image)


duration = int(1000 / fps)
#duration = 1750

# Save the frames as a GIF with the specified FPS
frames[0].save(output_file, format='GIF', append_images=frames[1:], save_all=True, duration=duration, loop=0,
               optimize=True, quality=100)

# frames_rgb = [frame.convert('RGB') for frame in frames]

# # Save the frames as a GIF with the specified FPS
# frames_rgb[0].save(output_file, format='GIF', append_images=frames_rgb[1:], save_all=True, duration=duration, loop=0,
#                    optimize=True, quality=100)
```
